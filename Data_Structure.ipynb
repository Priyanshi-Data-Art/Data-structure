```python
#1 Discuss string slicing and provide examples.

'''String slicing is used to access substring of a string. In python we can access some specific part of string using string slicing.
The Syntax of string slicing is:-

String[start:end:step]

where, start -> Represent the index where slice start. The index of start is inclusive.
       end -> Represent the index where slice ends. The index of end is exclusive. 
       step -> Represent the number of step move between character.
        
        
Example -> 
'''
str1 = "I can do coding"
print(str1[9:15]) # output coding

'''if we want to skip end index -> it means slice will go to starting index(that is define) to end of string.'''
print(str1[9:])

'''if we want to skip start index -> it means slice starts from begining of string and go to end index.'''
print(str1[:5])

'''As we see above, we have not give step, so by defalut it consider as 1. But if we define step as well then let's see how it's work.'''
print(str1[::2]) #output Icnd oig
print(str1[::-1]) #output gnidoc od nac I

'''In first case it skip one character and print another. Move two steps together.
In second case it starts from -1 index and goes till last.'''

```

```python
#2 Explain the key features of lists in Python.
'''Key feature of list in python:-

Ordered Collection:
Lists in Python maintain the order of elements. The position of each element is fixed and can be accessed by its index.'''
my_list = [10, 20, 30]
print(f"To see the 1 index of my_list :- {my_list[1]}")  

'''Mutable:
Lists are mutable, meaning you can change the elements (add, modify, or remove) after the list has been created.'''
my_list[1] = 40
print(f"Now list has bees changed {my_list}")  

'''Heterogeneous:
Lists can store elements of different data types (integers, strings, floats, even other lists).'''
my_list = [10, "Hello", 3.14, [1, 2, 3]]
print(my_list) 

'''Indexed Access:
Each element in a list can be accessed using its index. Python lists use zero-based indexing.'''
print(f"Here is the 0 index element {my_list[0]}") 

'''Duplicate Elements Allowed:
Unlike sets, Lists can contain duplicate elements.'''
my_list = [10, 20, 10, 30]
print(f"Here is the list with duplicat element {my_list}") 


```

```python
#5 Describe the key features of sets and provide examples of their use.

'''Key Featur of sets in Python:-


->Unordered Collection:
Sets are used to store the data in unordered manner, it is not allowed indexing.'''
sets={1,2,6,3}
print(f"May be the order of output will be change {sets}")


'''
->No Duplicate Elements:
Sets automatically remove any duplicate elements. This is one of their most useful features when you want to ensure uniqueness.'''
my_set = {10, 20, 10, 30}
print(f"It will automatically remove duplicacy {my_set}.")


'''
->Mutable:
While the elements in a set must be immutable (e.g., numbers, strings, tuples), the set itself is mutable. 
You can add or remove elements from a set.'''
my_set = {10, 20, 30}
my_set.add(40)
print(my_set)  # Output: {10, 20, 30, 40}


'''
->Set Operations (Union, Intersection, Difference):
Sets provide built-in methods for common set operations like union, intersection, and difference, which are
very efficient compared to lists.'''

set1 = {1, 2, 3}
set2 = {3, 4, 5}
print(set1&set2)         
print(set1 | set2)
print(set1-set2)   

'''
->Immutable Variant - frozenset:

Python also has an immutable version of sets called frozenset. Once created, elements in a 
frozenset cannot be modified (i.e., no add/remove).'''
my_frozenset = frozenset([10, 20, 30])
print(f"Now it's become immutable set {my_frozenset}") 

'''
Use-case:-
we use sets when we want to store unique items. In this case sets are the ideal option.
we can use sets if we want to perform some special mathematic method such as union, intersection and difference etc. 
'''

```

```python
#7 Describe how to add, modify, and delete items in a dictionary with examples.
'''Dictionary is mutable object that use to store elements in form of key-value pair.
syntax of dictionary:-'''
dic={"aman":1,"ajay":2,"akhil":3}
print(f"here is the dictionary {dic}")

'''As, it is mutable so we can do modification as well in dictionary. Let's play with dictionary.

-> first understand how to add item in dictionary.
In dictionary, we use key for adding an element unlike list. But we can not take duplicate key in dictionary that will thrown an error.
'''
dic["veer"] = 4
print(f"the new item has been add in dictionary {dic}")

'''
-> Now understand how to modify item in dictionary.
In this also we will use key to modify an value. for example:-'''
print(f"This is the old dictionary {dic}")
dic['aman']=100
print(f"This is the new dictionary {dic}")



'''
-> Now understand how to delete item in dictionary.
There is some method in python to remove dictionary. for example-> pop(key), popitem(), del

first use del keyword->'''
del dic['aman']
print(f"dictionary without aman {dic}")


'''Now use pop()->'''
dic.pop("ajay")
print(f"dictionary without ajay {dic}")


'''Now let's use popitem()'''
dic.popitem()
print(f"dictionary without value 3 {dic}")



```
